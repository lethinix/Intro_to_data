```python
# Visualize locations of interest by state
locations = [item["value"] for item in data["label_values"][0]["vec"]]

# Extract state names (text after the final comma when present).
states = [loc.split(",")[-1].strip() for loc in locations]
state_counts = pd.Series(states).value_counts().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
state_counts.plot(kind="bar", color="steelblue", edgecolor="black")
plt.title("Locations of Interest by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## Data exploration 2 - ad data from instagram

```python
import json

data_file2 = "/Users/aishakazembe/Documents/coding-cabinate/code/Intro_to_data/Intro_to_data/proj1/Project01-main/data/my_data/ads_information/ads_and_topics/ads_clicked.json"

with open(data_file2) as f:
    data2 = json.load(f)

print(data2)
```

```python
# Ads data visualization: which ad titles I clicked most often
import os

ads_df = pd.DataFrame(data2)

# Pull label/value pairs into normal columns for easier analysis.
def get_label_value(label_values, target_label):
    for item in label_values:
        if item.get("label") == target_label:
            return item.get("value", "Unknown")
    return "Unknown"

# Fix common mojibake so chart labels are readable.
def fix_mojibake(text):
    if not isinstance(text, str):
        return text
    try:
        return text.encode("latin1").decode("utf-8")
    except (UnicodeEncodeError, UnicodeDecodeError):
        return text

ads_df["title"] = ads_df["label_values"].apply(lambda x: get_label_value(x, "Title"))
ads_df["title"] = ads_df["title"].apply(fix_mojibake)
ads_df["action"] = ads_df["label_values"].apply(lambda x: get_label_value(x, "Action"))
ads_df["clicked_at"] = pd.to_datetime(ads_df["timestamp"], unit="s")

title_counts = ads_df["title"].value_counts().head(10)

plt.figure(figsize=(11, 5))
title_counts.plot(kind="bar", color="#2a9d8f", edgecolor="black")
plt.title("Most Clicked Ad Titles")
plt.xlabel("Ad Title")
plt.ylabel("Clicks")
plt.xticks(rotation=25, ha="right")
plt.tight_layout()

# Save chart image for part03.md deliverable.
os.makedirs("imgs", exist_ok=True)
plt.savefig("imgs/ads_clicks_by_title.png", dpi=200, bbox_inches="tight")
plt.show()

print("Saved plot to imgs/ads_clicks_by_title.png")
```
