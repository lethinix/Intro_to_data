```python
filenames = [
  "image0.jpg", "image1.JPG", "text.txt", "image4.png", 
  "image10.jpeg", "image11.JPEG", "archive.gz", "image14.webp",
]

for fname in filenames:
  if fname[-3:].lower() == "jpg":
    print("open " + fname)
    txt_fname = fname.lower().replace("jpg", "txt")
    print("writing out info to: " + txt_fname)
```

```python
# TODO: handle .png, .jpg, .jpeg, .webp, .gz, .qt

for fname in filenames:
  extension = fname.split(".")[-1].lower()
  if extension in ("png", "jpg", "jpeg", "webp", "bmp", "gif"):
    print("open and convert " + fname)
    txt_fname = ".".join(fname.split(".")[:-1]) + ".png"
    print("writing out info to: " + txt_fname)
```

```python
fname = "image11.JPEG"
fname.split(".")
```

```python
# TODO: inverse split
name_parts = fname.split(".")
".".join(name_parts[:-1]) + ".png"
```

```python
# TODO: write text file
processed_files = []

for fname in filenames:
  extension = fname.split(".")[-1].lower()
  if extension in ("png", "jpg", "jpeg", "webp", "bmp", "gif"):
    processed_files.append(".".join(fname.split(".")[:-1]) + ".png")

processed_from_file = "\n".join(processed_files)

with open("processed.txt", "w") as wfile:
  wfile.write(processed_from_file)

processed_from_file
```
